```python
days= [1,2,3,4,5,6,7] 
max_temp=[45,51,52,48,47,49,46] 
min_temp=[43,42,40,44,33,35,37] 
avg_temp=[49,48,48,46,40,42,41] 
plt.plot(days, max_temp, label="max") 
plt.plot(days, min_temp, label="min") 
plt.plot(days, avg_temp, label="average") 
plt.xlabel('Day') 
plt.ylabel('Temperature') 
plt.title('Weather') 
plt.legend() 
plt.show() 
```

```python
plt.plot(days, max_temp, label="max") 
plt.plot(days, min_temp, label="min") 
plt.plot(days, avg_temp, label="average") 
plt.xlabel('Day') 
plt.ylabel('Temperature') 
plt.title('Weather') 
plt.legend(loc='upper left',shadow=True,fontsize='medium') 
plt.show() 
```

```python
data = 'Python', 'C++', 'R', 'Swift' 
sizes = [200, 130, 205, 215]
clr = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'] 
explode = (0, 0.1, 0, 0) 
plt.pie(sizes, explode=explode, labels=data, colors=clr, autopct='%1.1f%%') 
plt.axis('equal') 
plt.show()
```

```python
company=['C1', 'C2', 'C3', 'C4'] 
revenue=[90,136,89,27] 
plt.bar(company, revenue, label="Revenue") 
plt.ylabel("Revenue") 
plt.title('US Technology Stocks') 
plt.legend() 
plt.show()
```

```python
plt.barh(company, revenue, label="Revenue") 
plt.ylabel("Revenue") 
plt.title('US Technolog Stocks') 
plt.legend()
```

```python
profit= [40,2,34,12] 
plt.bar(company, revenue, label="Revenue") 
plt.bar(company, profit, label="Profit") 
plt.title('US Technology Stocks') 
plt.legend() 
plt.show() 
```

```python
xpos = np.arange(len(company)) 
plt.bar(xpos-0.2, revenue, width=0.4, label="Revenue") 
plt.bar(xpos+0.2, profit, width=0.4, label="Profit") 
plt.xticks(xpos, company) 
plt.ylabel("Revenue") 
plt.title('US Technology Stocks') 
plt.legend() 
plt.show() 
```
